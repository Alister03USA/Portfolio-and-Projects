# Predicting total customer purchased based on demographics (Income, Age, Marital_Status, Education) using RandomForest

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Define features (X) and target (y)
x = df[['Income', 'Age', 'Marital_Status', 'Education']]
y = df['Total Purchased']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

# Define preprocessing steps
numeric_features = ['Income', 'Age']
categorical_features = ['Marital_Status', 'Education']

# Create preprocessing pipeline
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(drop='first',handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])

# Append RandomForestRegressor to preprocessing pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor())
])

# Train the model (including preprocessing)
pipeline.fit(X_train, y_train)

# Make predictions
y_pred = pipeline.predict(X_test)

# Create a scatter plot of the actual vs predicted values
plt.scatter(y_test, y_pred, edgecolor = "black", color = 'lightblue')

# Add a diagonal line to represent perfect prediction
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'k--', lw=2)

# Label the axes
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")

# Show the plot
plt.show()

# Find the mean square error, mean absolute error and R^2 of the model
mse = mean_squared_error(y_test, y_pred) #Calculates the average squared difference between the actual and predicted values
print("Mean Squared Error:", mse)

mae = mean_absolute_error(y_test, y_pred) # Calculate the average absolute difference between actual and predicted values
print("Mean Absolute Error:", mae)

r2 = r2_score(y_test, y_pred) # Measure how well the model fits the data
print("R^2: ", r2)



#Cross Validation of the model
cv_scores_rf = cross_val_score(pipeline, x, y, cv=5, scoring='neg_mean_squared_error')

# Convert scores from negative MSE to MSE
cv_scores_rf = -cv_scores_rf

# Print cross-validation scores
print("Cross-Validation Scores (RandomForestRegressor):")
print(cv_scores_rf)

# Calculate mean and standard deviation of cross-validation scores
print("Mean MSE:", cv_scores_rf.mean())
print("Standard Deviation MSE:", cv_scores_rf.std())
```

# Conclusion:

The mean MSE of approximately 111720 indicates the average squared difference between the predicted values and the true values of the target variable (total purchased amount) across the validation sets.
A lower MSE value indicates better model performance, as it represents smaller errors between predicted and actual values.

In summary, the RandomForestRegressor model, when evaluated using 5-fold cross-validation on the dataset, has an average Mean Squared Error (MSE) of 111720 and a standard deviation of approximately 11481. This information helps assess the model's performance and provides insights into the consistency of its predictions across different subsets of the data. Lower MSE values and lower variability (standard deviation) generally indicate better model performance and stability.

# Predicting total customer purchased based on demographics (Income, Age, Marital_Status, Education) using MLPRegressor

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

# Define features (X) and target (y)
X = df[['Income', 'Age', 'Marital_Status', 'Education']]
y = df['Total Purchased']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Define preprocessing steps
numeric_features = ['Income', 'Age']
categorical_features = ['Marital_Status', 'Education']

# Create preprocessing pipeline
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(drop='first',handle_unknown='ignore'))  # Use OneHotEncoder with drop='first' to handle categorical data
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])

# Append MLPRegressor to preprocessing pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', MLPRegressor(hidden_layer_sizes=(100, 50), activation='relu', solver='adam', max_iter=1000, random_state=1))
])

# Train the model (including preprocessing)
pipeline.fit(X_train, y_train)

# Make predictions
y_pred = pipeline.predict(X_test)

# Create a scatter plot of the actual vs predicted values
plt.scatter(y_test, y_pred, edgecolor="black", color = 'red')

# Add a diagonal line to represent perfect prediction
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], 'k--', lw=2)

# Label the axes
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")

# Show the plot
plt.show()

# Find the mean square error, mean absolute error, and R^2 of the model
mse = mean_squared_error(y_test, y_pred)  # Calculates the average squared difference between the actual and predicted values
print("Mean Squared Error:", mse)

mae = mean_absolute_error(y_test, y_pred)  # Calculate the average absolute difference between actual and predicted values
print("Mean Absolute Error:", mae)

r2 = r2_score(y_test, y_pred)  # Measure how well the model fits the data
print("R^2: ", r2)



#Cross Validation of the model
cv_scores_rf = cross_val_score(pipeline, x, y, cv=5, scoring='neg_mean_squared_error')

# Convert scores from negative MSE to MSE
cv_scores_rf = -cv_scores_rf

# Print cross-validation scores
print("Cross-Validation Scores (RandomForestRegressor):")
print(cv_scores_rf)

# Calculate mean and standard deviation of cross-validation scores
print("Mean MSE:", cv_scores_rf.mean())
print("Standard Deviation MSE:", cv_scores_rf.std())
```

# Conclusion

**Interpretation:**

The mean MSE of approximately 100240 indicates the average squared difference between the predicted values and the true values of the target variable (total purchased amount) across the validation sets.
A lower MSE value indicates better model performance, as it represents smaller errors between predicted and actual values.

**Insights:**

The mean MSE of 100240 suggests that, on average, the MLPRegressor model is predicting the total purchased amount with a squared error of approximately 100240 units (e.g., currency squared).
The relatively low standard deviation of 8918 indicates that the model's performance does not varies as significant, suggesting some consistency in prediction quality.


In summary, while the MLPRegressor model is able to predict the total purchased amount to a certain extent, there is room for improvement, especially in reducing the variability of predictions across different subsets of the data.
